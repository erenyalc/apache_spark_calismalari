```python
sdf = sdf.dropna()
```

```python
sdf = sdf.withColumn('age_kare', sdf.age**2)
sdf.show(10)
```

```python
from pyspark.ml.feature import StringIndexer
strIndexer = StringIndexer(inputCol = 'churn', outputCol = 'label')
#bağımlı değişkenlerin ayarlanması
```

```python
indexed = strIndexer.fit(sdf).transform(sdf)
```

```python
indexed.dtypes
```

```python
sdf = indexed.withColumn('label', indexed['label'].cast('integer'))
```

```python
from pyspark.ml.feature import VectorAssembler
sdf.columns
```

```python
#bağımsız değişkenlerin ayarlanması
bag = ['age','total_purchase','account_manager', 'years','num_sites']
```

```python
vectorassambler = VectorAssembler(inputCols = bag, outputCol = 'features')
```

```python
va_df = vectorassambler.transform(sdf)
```

```python
fsdf = va_df.select(['features','label'])
```

```python
fsdf.show(10)
#spark ml işlemleri için belli bir format ister. Bunlar features ve label gibi bir yapıdır. 
#Bağımlı ve bağımsız değişkenleri bu şekilde features ve label'a çevirirsek işimizi kolaylaştırmış oluruz.
```

```python
#test-train
```

```python
splits = fsdf.randomSplit([0.7,0.7])
train_df = splits[0]
test_df = splits[1]
```

```python
train_df.show(5)
```

```python
test_df.show(5)
```

```python
#gbm ile müşteri terk modellemesi
```

```python
from pyspark.ml.classification import GBTClassifier
```

```python
gbm = GBTClassifier(maxIter = 10, featuresCol = 'features', labelCol = 'label')
gbm_model = gbm.fit(train_df)
```

```python
y_pred = gbm_model.transform(test_df)
```

```python
ac = y_pred.select('label','prediction')
```

```python
ac.filter(ac.label == ac.prediction).count() / ac.count()
#ilk test hatası
```

```python
#model tuning
from pyspark.ml.evaluation import BinaryClassificationEvaluator
from pyspark.ml.tuning import ParamGridBuilder, CrossValidator

evaluator = BinaryClassificationEvaluator()

paramGrid = (ParamGridBuilder()
           .addGrid(gbm.maxDepth, [2,4,6])
           .addGrid(gbm.maxBins, [20,30])
           .addGrid(gbm.maxIter, [10,20])
           .build())
cv = CrossValidator(estimator = gbm, estimatorParamMaps = paramGrid, evaluator=evaluator, numFolds= 10)
```

```python
cv_model = cv.fit(train_df)
```

```python
y_pred = cv_model.transform(test_df)
ac = y_pred.select('label','prediction')
```

```python
ac.filter(ac.label == ac.prediction).count() / ac.count()
```

```python
evaluator.evaluate(y_pred)
```
